```python
df= data.copy()
```

```python
df.head()
```

```python
df2 = df.groupby(['Country'])[['Confirmed','Recovered','Deaths']].max().reset_index()
```

```python
import geopy
from geopy.geocoders import Nominatim
lat =[]
lon =[]
geolocator=Nominatim(user_agent='app')
for location in df2['Country']:
  location =geolocator.geocode(location)
  if location is None:
    lat.append(np.nan)
    lon.append(np.nan)
  else:
    lat.append(location.latitude)
    lon.append(location.longitude)
```

```python
df2['lat']=lat
df2['lon']=lon
```

```python
df2.head()
```

```python
! pip install folium
```

```python
import folium   #MArkers
m=folium.Map(location=[66,15],zoom_start=2)
df2=df2.dropna(subset=['lat'])
df2=df2.dropna(subset=['lon'])
for id,row in df2.iterrows():
  row['lat'] = pd.to_numeric(row['lat'])
  row['lon'] = pd.to_numeric(row['lon'])
  folium.Marker(location=[row['lat'],row['lon']],popup=folium.Popup(row['Confirmed'])).add_to(m)
m
```

```python
#Geographic heat map to show low density and high density regions of Confirmed cases
from folium.plugins import HeatMap
n=folium.Map(location=[64,19],tiles='openstreetmap',zoom_start=2)
HeatMap(data=df2[['lat','lon','Confirmed']].to_numpy(),radius=15).add_to(n)
n
```
